```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def pltImage(img):
    channels = len(img.shape)
    if channels == 3:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap='gray')
    plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis
    plt.show()
```

```python
img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[237,269],[570,230],[566,496],[235,462]])
pts2 = np.float32([[200,200],[500,200],[500,400],[200,400]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(1024,768))
plt.rcParams["figure.figsize"] = (20,10)
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('book.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[350,360],[800,370],[30,880],[760,1000]])
pts2 = np.float32([[0,0],[800,0],[0,800],[700,800]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(958,1276))
plt.rcParams["figure.figsize"] = (20,10)
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
